```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# checking for missing values
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)
```

```python
import numpy as np
```

```python
books['missing_description'] = np.where(books['description'].isna(),1,0)
books['age_of_book'] = 2025 - books['published_year']
```

```python
columns_of_interest = ['num_pages', 'age_of_book','missing_description','average_rating']
correlation_matrix = books[columns_of_interest].corr(method = 'spearman')

sns.set_theme(style ='white')
plt.figure(figsize=(8,6))
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm',fmt='.2f', cbar_kws={'label':'correlation matrix'})
heatmap.set_title('correlation matrix')
plt.show()
```

```python
clean_books=books[~(books['description'].isna())&
      ~(books['num_pages'].isna())&
      ~(books['average_rating'].isna())&
      ~(books['published_year'].isna())].copy()
```

```python
clean_books['categories'].value_counts().reset_index().sort_values("count", ascending = False)
```

```python
clean_books['words_in_description'] = clean_books['description'].str.split().str.len()
```

```python
clean_books.loc[clean_books['words_in_description'].between(25,34),'description']
```

```python
clean_books_25_words = clean_books[clean_books["words_in_description"] >=25]
```

```python
clean_books_25_words
```

```python
clean_books_25_words = clean_books_25_words.copy()
```

```python
clean_books_25_words['title and subtitle'] = (np.where(clean_books_25_words['subtitle'].isna(), clean_books_25_words['title'],
                                                      clean_books_25_words[['title','subtitle']].astype(str).agg(": ".join,axis=1)))
```

```python
clean_books_25_words
```

```python
clean_books_25_words['tagged_description'] = clean_books_25_words[['isbn13', 'description']].astype(str).agg(" ".join,axis=1)
```

```python
clean_books_25_words
```

```python
(clean_books_25_words
.drop(['subtitle','words_in_description', 'age_of_book','missing_description'], axis=1)
.to_csv('books_cleaned.csv'),index = False
)
```
